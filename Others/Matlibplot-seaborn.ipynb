```python
sns.scatterplot(data=df,x='engine_size',y='horsepower');
```

```python
sns.lmplot(data=df,x='curb_weight',y='horsepower',col='fuel_type',ci=False);
```

```python
sns.violinplot(data=df,x="body_style",y='engine_size',orient='v',palette='colorblind');
```

```python
plt.figure(figsize=(10,7));
sns.stripplot(data = df, x = 'number_of_doors', y='horsepower',jitter=True);
```

```python
sns.swarmplot(data = df, x = 'fuel_type', y='price',hue='number_of_doors',dodge=True);
```

```python
sns.catplot(data = df, x = 'fuel_type', y='price',kind='swarm');
```

```python
sns.pairplot(data = df, vars = ['normalized_losses','wheel_base','curb_weight','engine_size','peak_rpm'],hue='number_of_doors',corner=True);
```

```python
plt.xticks(rotation=30)
sns.heatmap(data = df[['normalized_losses','wheel_base','curb_weight','engine_size','peak_rpm']].corr(), annot=True,cbar=False);
```

```python
import plotly.express as px
bar = px.scatter_3d(df,x='fuel_type',y='horsepower',z='price',color='horsepower')
bar.show()
```
